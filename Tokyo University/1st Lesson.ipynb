# 時系列の前処理と可視化(1)

```python
# データのインポート
import pandas as pd

def data_read():

    # 船舶データ
    hakusan = pd.read_csv("./data/hakusan_new.csv", encoding = 'shift-jis')
    # 太陽黒点数データ
    sunspot = pd.read_csv("./data/sunspot_new.csv", encoding = 'shift-jis')
    # 東京の日最高気温データ
    maxtemp = pd.read_csv("./data/maxtemp.csv", encoding = 'shift-jis')
    # アメリカの食品産業に従事する労働者人口
    blsfood = pd.read_csv("./data/blsfood_new.csv", encoding = 'shift-jis')
    # 工業製品の卸売高
    whard = pd.read_csv("./data/whard_new.csv", encoding = "shift-jis")
    # 地震データ
    mye1f = pd.read_csv("./data/mye1f_new.csv", encoding = "shift-jis")
    # 日経255データ
    nikkei225 = pd.read_csv("./data/nikkei225_new.csv", encoding = "shift-jis")
    # 東京降雨データ（２年間の降雨日数）
    rainfall = pd.read_csv("./data/rainfall_new.csv", encoding = "shift-jis")
    # 榛原地下水位データ
    haibara = pd.read_csv("./data/haibara_new.csv", encoding = "shift-jis")
    # カナディアンリンクスの捕獲数
    lynx = pd.read_csv("./data/Lynx-new.csv")

    return hakusan, sunspot, maxtemp, blsfood, whard, mye1f, nikkei225, rainfall, haibara, lynx
```

```python
hakusan, sunspot, maxtemp, blsfood, whard, mye1f, nikkei225, rainfall, haibara, lynx = data_read()
```

```python
# 方角速度のデータを可視化
## 定常性 / 周期性
hakusan["方向角速度"].plot()
```

```python
# 太陽黒点数データ
## 正値・上下非対称 / 擬似周期性 / 前後非対称性
sunspot.plot()
```

```python
# 日最高気温データ（東京）
## トレンド（長期の周期性） / トレンド周りほぼ定常
maxtemp.plot()
```

```python
# 米国食品産業従事者数データ
## 年周期性 / トレンド
blsfood.plot()
```

```python
# 卸売高データ
## 年周期性 / 生値性 / トレンドと変動分散の連動
whard.plot()
```

```python
# MYE1F(東西方向地震波)データ
## トレンドなし / 分散非定常・共分散非定常 / 区分的定常
mye1f.plot()
```

```python
# 日経２２５平均株価データ
## トレンド＋分散変動 / トレンド下降時の分散増加
nikkei225.plot()
```

```python
# Canadian Lynx 捕獲数
## 平均定常 / 擬似周期的 / 前後(左右)非対称
lynx.plot()
```

```python
# 榛原地下水位・気圧データ
## ２変量データ・逆相関 / 異常値・欠測値を含む
import matplotlib.pyplot as plt
import japanize_matplotlib

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(haibara["地下水位"], label = "地下水位")
ax1.legend()
ax2.plot(haibara["気圧"], label = "気圧")
ax2.legend()
plt.show()
```
